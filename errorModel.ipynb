```python
import numpy as np
import matplotlib.pyplot as plt

result_path = 'results_object.txt'

exList = []
eyList = []
dList = []
with open(result_path, 'r') as f:
    lines = f.readlines()
    for line in lines[1:]:
        line = line.strip()
        ex = line.split()[1]
        ey = line.split()[2]
        x = line.split()[3]
        y = line.split()[4]
        d = np.linalg.norm([float(x), float(y)])
        exList.append(float(ex))
        eyList.append(float(ey))
        dList.append(d)

# plot hist with 100 bins
plt.hist(exList, bins=100, color='blue', alpha=0.7)
plt.xlim(-1,1)
plt.xlabel('Error in x direction')
plt.ylabel('Frequency')
plt.title('Error histogram')
plt.show()

plt.hist(eyList, bins=100, color='red', alpha=0.7)
plt.xlabel('Error in y direction')
plt.ylabel('Frequency')
plt.title('Error histogram')
plt.show()
```

```python
print("std in x direction: ", np.std(exList))
print("std in y direction: ", np.std(eyList))
print("coefficient of ex ey: ", np.corrcoef(exList, eyList)[0,1])
```
